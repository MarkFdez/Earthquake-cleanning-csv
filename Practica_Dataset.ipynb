```python

import pandas as pd
import re


df = pd.read_csv('earthquake.csv')

import pandas as pd

df = pd.read_csv('earthquake.csv')

configuracion = [
    ("Unknown Number of Deaths", "Unknown Number of Deaths, lat", "Unknown Number of Deaths, lon"),
    ("0 Deaths", "0 Deaths, lat", "0 Deaths, lon"),
    ("1-50 Deaths", "1-50 Deaths, lat", "1-50 Deaths, lon"),
    ("51-100 Deaths", "51-100 Deaths, lat", "51-100 Deaths, lon"),
    ("101-1000 Deaths", "101-1000 Deaths, lat", "101-1000 Deaths, lon"),
    (">1001 Deaths", ">1001 Deaths, lat", ">1001 Deaths, lon")
]

lista_final = []

for col_lugar, col_lat, col_lon in configuracion:

    sub_tabla = df[[col_lugar, col_lat, col_lon]].copy()

    sub_tabla.columns = ['Lugar', 'Latitud', 'Longitud']

    sub_tabla['Rango_Muertes'] = col_lugar

    sub_tabla = sub_tabla.dropna(subset=['Lugar'])

    lista_final.append(sub_tabla)

df_resultado = pd.concat(lista_final, ignore_index=True)

print("Datos reestructurados con relación directa:")
display(df_resultado)

```
